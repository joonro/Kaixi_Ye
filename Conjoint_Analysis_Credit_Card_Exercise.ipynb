```python
# -*- coding: utf-8 -*-

from __future__ import division 

import os  # basic OS module

# import scientific libraries
from numpy import *  # numpy for fundamendal n-dimensional array datatype
import matplotlib.pyplot as plt  # matplotlib.pyplot for plots

import pandas as pd  # pandas for DataFrame
import statsmodels.formula.api as smf  # statsmodels for regression using formula
```

```python
# load the data using pandas

# pandas.read_table() method

conjoint = pd.DataFrame.from_csv("Credit_Card_Conjoint_All.csv")

N = conjoint.shape[0] - 1

#print(conjoint.columns)

# identify questions
ix_start = None
J = 0
for j, col in enumerate(conjoint.columns):
    if col.find("id") != -1:
        if ix_start is None:
            ix_start = j
        
        J += 1

# generate a DataFrame y, that consists of ratings
y = pd.DataFrame(data=conjoint.iloc[1:, ix_start:ix_start + J].astype(int))
#conjoint
```

### Generate the design matrix for Conjoint analysis


```python
from collections import OrderedDict

# Dictionary to store attributes and labels
levels_per_attr = OrderedDict()
levels_per_attr["Brand"] = ["Mastercard", "Amex", "Visa"]
levels_per_attr["Tier"] = ["Regular", "Gold", "Platinum"]
levels_per_attr["Annual Fee"] = ["$0 Annual Fee",
                                 "$10 Annual Fee",
                                 "$25 Annual Fee"]
levels_per_attr["Cashback"] = ["No Cashback", "1% Cashback"]
levels_per_attr["APR"] = ["0% APR", "10% APR"]
levels_per_attr["Credit Limit"] = ["$1,000 Credit Limit",
                                   "$5,000 Credit Limit",
                                   "$10,000 Credit Limit"]

X = pd.DataFrame()


for i, attribute in enumerate(levels_per_attr.keys()):
    levels = levels_per_attr[attribute]

    for k, level in enumerate(levels):
        if k:  # if k > 0
            X[level] = zeros((J, ), dtype=int)

```

### Populate the design matrix with the survey data

```python
ix = 0
    
for col, question in conjoint.iloc[0, :].iteritems():
    if col.find("id") != -1:
        for i, x in enumerate(X.columns):
            if x in question:
                X.loc[ix, x] = 1
        ix += 1
X    
```

### Individual Level Parameters

Run one regression per respondants

### Separate regression for each individual

```python

# loop over respondents
x_names = list(X.columns)
ols_x_names = []
for i in range(len(x_names)):
    ols_x_names.append("x" + str(i + 1))
    
formula = 'y ~ x1 + x2 + x3 + x4 + x5 + x6 + x7 + x8 + x9 + x10'

data_ind = pd.DataFrame(data=X.values, columns=ols_x_names)

for i in range(N):
    data_ind['y'] = y.iloc[i].values
    
    OLS = smf.ols(formula, data=data_ind)
    results_ind = OLS.fit()

    if i == 0:
        partworths = pd.DataFrame(data=results_ind.params.reshape(1, -1),
                                  columns=['Intercept'] + list(X.columns))
    else:
        partworths.loc[i, :] = results_ind.params.reshape(1, -1)
        
#partworths    
```

```python
partworths=partworths.rename(columns={'Amex':'Brand_Amex','Visa':'Brand_Visa','Gold':'Tier_Gold','Platinum':'Tier_Platinum'})#['Brand_Amex','Brand_Visa','Tier_Gold','Tier_Platnum']
```

## Individual importance weight for each attribute
The individual importance weight for each attribute is suspect to great variation as can be seen in following boxpolts: First, the weight of the Credit Limit is much higher than the reminding attributes, and the weight of brand is the lowest. Second,Annual Fee, Cashback and ARP has the similar weights for this sample.Thierd, the variance for credit limit is very big, which means the importance fr credit limit is not stable.

```python
ranges = pd.DataFrame()  # DataFrame to hold ranges
Factors=['Brand','Tier','Annual Fee','Cashback','APR','Credit Limit']
for i, attribute in enumerate(Factors):

    if attribute == "Brand":
        tmp_max = maximum(partworths["Brand_Amex"], partworths["Brand_Visa"])
        attribute_max = maximum(0, tmp_max)

        tmp_min = minimum(partworths["Brand_Amex"], partworths["Brand_Visa"])
        attribute_min = minimum(0, tmp_min)

    elif attribute == "Tier":
        tmp_max = maximum(partworths["Tier_Gold"], partworths["Tier_Platinum"])
        attribute_max = maximum(0, tmp_max)

        tmp_min = minimum(partworths["Tier_Gold"], partworths["Tier_Platinum"])
        attribute_min = minimum(0, tmp_min)

    # In this case it is yes or no, so you only need to compare to 0
    elif attribute == "Annual Fee":
        attribute_max = maximum(partworths["$10 Annual Fee"], partworths["$25 Annual Fee"])
        attribute_min = minimum(partworths["$10 Annual Fee"], partworths["$25 Annual Fee"])

    # TODO: calculate range for the Expansion Slot attribute
    elif attribute == "Cashback":
        attribute_max = maximum(0, partworths["1% Cashback"])
        attribute_min = minimum(0, partworths["1% Cashback"])
    # TODO: calculate range for the price attribute     
    elif attribute == "APR":
        attribute_max = maximum(0, partworths["10% APR"])
        attribute_min = minimum(0, partworths["10% APR"])
    elif attribute == "Credit Limit":
        tmp_max = maximum(partworths["$5,000 Credit Limit"], partworths["$10,000 Credit Limit"])
        attribute_max = maximum(0, tmp_max)
        tmp_min = minimum(partworths["$5,000 Credit Limit"], partworths["$10,000 Credit Limit"])
        attribute_min = minimum(0, tmp_min)
    
    # store the range
    ranges[attribute] = attribute_max - attribute_min
      
importance = ranges.copy()
importance_sum = importance.sum(axis=1)

# divide each element with the corresponding sum along with axis 0
importance = importance.div(importance_sum, axis=0)  
#importance
```

##Cluster Analysis
From the plot of cluster analysis, we can see the preference on attributes are different between groups. On average people value credit limit as the most important attribute.However, the biggest group ,group 3, didn't represent the same preference, they value cashback the most. And the smaller group,group2, value the credit limit as a decisive attribute. For the smallest group, group1, they value APR and Annual Fee more than other 2 groups.

```python
from sklearn.cluster import KMeans

importance3=importance.copy()
importance4=importance.copy()
# Initialize an K-means object with 3 and 4 cluster centers
kmeans_3 = KMeans(n_clusters=3)
kmeans_4 = KMeans(n_clusters=4)
# Run K-means clustering and store the segment membership
segment3=kmeans_3.fit(importance).labels_
segment4=kmeans_4.fit(importance).labels_
#importance["segment"] = kmeans_5.fit(importance).labels_

# pandas will automatically figure out the correct obs
importance3['segment']=segment3
importance4['segment']=segment4
#importance

```
